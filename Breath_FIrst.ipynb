### **TASK01 - Breadth Search Using Python**
#### Summary
This task is part of MSc AI Hull Task Assignment for Week03 Course about Problem Solving by Search to solve path search by Breadth Search Algorithm

**1. Import necessary library --> igraph and matplotlib**

```python
import igraph as ig
g = ig.Graph()
import matplotlib.pyplot as plt
```

**2. Create vertice and edges representing problematic model and plot the graph**

```python
g.add_vertices(8)
g.vs['label'] = ['0','1','2','3','4','5','6','7']
g.add_edges([(0,7),(0,5),(0,3),(7,2),(7,4),(5,1),(3,6)])
# g.vs['size'] = 30
```

```python
ax = plt.subplot()
ig.plot(g, target = ax,edge_color = 'gray',vertex_color = 'orange',vertex_size = 30) 
```

**3. Implement BFS Search and print the result**

We got this path [0, 3, 5, 7, 6, 1, 2, 4]

```python
bfs_order = g.bfs(0)
order,parents, distances = bfs_order
print("BFS Order",order)
print("Parents", parents)
print("DIstances", distances)
```

**4. Drawing the BFS result on graph to be overlayed with original graph**

```python
g2 = ig.Graph(directed = True)
g2.add_vertices(8)
g2.add_edges([(0,3),(3,5),(5,7),(7,6),(6,1),(1,2),(2,4)])
g2.vs['label'] = ['0','1','2','3','4','5','6','7']
print(g2)
```

```python
ax = plt.subplot()
ig.plot(g2, target = ax,edge_color = 'gray',vertex_color = 'orange',vertex_size = 30,vertex_label = g2.vs['label'])
```

**5. Drawing Solution overlayed with original graph**

```python
ax = plt.subplot()
ig.plot(g, target = ax,edge_color = 'gray',vertex_color = 'orange',vertex_size = 30) 
layout = g.layout("kamada_kawai")
ig.plot(g2, layout =layout,target = ax,edge_color = 'gray',vertex_color = 'orange',vertex_size = 30)
```
