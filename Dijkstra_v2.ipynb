**2. Create vertice and edges representing problematic model and plot the graph**

```python
g5 = ig.Graph(directed=False)
g5.add_vertices(['A', 'B', 'C', 'D', 'E', 'F'])
g5.vs['label'] = ['A', 'B', 'C','D','E','F']
g5.add_edges([('A','B'),('A','D'),('B','D'),('B','E'),('C','E'),('C','F'),('D','E'),('E','F'),('F','D')])
weights_g5 = [2,8,5,6,9,3,3,3,2]
g5.es['weight'] = [2,8,5,6,9,3,3,3,2]
g5.es['label'] = [2,8,5,6,9,3,3,3,2]
ax = plt.subplot()
ig.plot(g5,target = ax,layout = 'kk',vertex_size = 30,vertex_color ='orange',edge_label = g5.es['label'])   
```
